```python
import pandas as pd
import re
```

```python
df = pd.read_csv('all_csv/All_in_one.csv')
df
```

```python
# Function to clean each message
def clean_text(text):
    # Keep only Cyrillic letters, numbers, and spaces
    cleaned_text = re.sub(r'[^А-Яа-яӘәІіҢңҒғҚқҰұҮүҺһӨөЁё0-9\s]', '', text)
    # Remove extra spacesa
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()
    return cleaned_text

# Apply the cleaning function and filter out empty messages
df['message'] = df['message'].apply(clean_text)
df = df[df['message'] != ""]
# df = df[~df['message'].str.startswith("Извините")]

# Filter out rows with fewer than 5 words
df = df[df['message'].apply(lambda x: len(x.split()) >= 5)]
df
```

```python
df.to_csv('all_csv/All_in_one.csv')
```
